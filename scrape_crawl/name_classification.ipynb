# Load puplications, get authors and how many publications of theirs we have

```python
from collections import Counter
```

```python
data = pd.read_csv("bb_pulications.csv")
data['author'] = data['bib'].apply(lambda x: ast.literal_eval(x)['author'])
data['year'] = data['bib'].apply(lambda x: ast.literal_eval(x)['pub_year'])
data['title'] = data['bib'].apply(lambda x: ast.literal_eval(x)['title'])
```

Clean author names a little bit

```python
def clean_author(x):
    """
    x list of authors for a publication
    """
    clean_list = []
    for item in x:
        clean_list.append(item.lower())
    return clean_list
```

```python
data['author_cleaned'] = data['author'].apply(lambda x: clean_author(x))
```

```python
authors = data['author_cleaned'].sum()
```

```python
Counter(authors)
```

Get surnames

```python
# first have to searate from initials and then put back togetehr again
surnames = []
for name in authors:
    surname = name.split(' ')[1:]
    surname_str = ' '.join(surname) 
    surnames.append(surname_str)
```

```python
Counter(surnames)
```

# Hokay, lets try this Classifier

```python
#drop duplicates
surnames = list(set(surnames))
```

```python
df = pd.DataFrame(surnames, columns=["surnames"])
```

```python
preds = pred_wiki_ln(df, "surnames")
```
