```python
def citations_via_url(result):
    dictor = {}
    print("iteration")
    pause = random.randint(0,40)
    print("pausing for {} seconds".format(pause) )
    time.sleep(pause)
    try:
        dictor = result
        citers_gen = scholarly.citedby(result)
        #we are looking at a max of 20 citations per article, I am a bit worried about being cutoff by google
        citers_list = []
        for i in range(20):
            try:
                pause = random.randint(0,15)
                #print("pausing for {} seconds".format(pause) )
                time.sleep(pause)
                citer = next(citers_gen)
                citers_list.append(citer['bib'])
            except StopIteration:
                print("At the end of the iteration. Total number of cited is {}".format(i))
                break
        dictor['citations'] = citers_list
        return dictor
    except KeyError:
        dictor = result
        dictor['citations'] = []
        print("No citations found for this article")
        return dictor
```

```python
def get_citations(sub_list):
    #takes in a list of dictionaries and returns the citations inside those dictionaries
    citations = []
    print(len(sub_list))
    for result in sub_list:
        citations.append(citations_via_url(result))
    return citations
```
