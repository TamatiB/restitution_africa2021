# Get those that have cited each article

```python
from collections import Counter
```

```python
years = []
for item in res:
    years.append((item['bib']['pub_year']))
```

```python
Counter(years)
```

```python
#from 2016
frm_2016 = []
for item in res:
    try:
        if int(item['bib']['pub_year']) >=2016:
            frm_2016.append(item)
    except ValueError:
        pass
```

```python
citations = []
for result in frm_2016:
    dictor = {}
    try:
        dictor = result
        citers_gen = scholarly.citedby(result)
        #we are looking at a max of 20 citations per article, I am a bit worried about being cutoff by google
        citers_list = []
        for i in range(20):
            try:
                citer = next(citers_gen)
                citers_list.append(citer['bib'])
            except StopIteration:
                print("At the end of the iteration. Total number of cited is {}".format(i))
                break
            
        dictor['citations'] = citers_list
        citations.append(dictor)
    except KeyError:
        dictor = result
        dictor['citations'] = []
        citations.append(dictor)
        print("No citations found for this article")
        pass
```
